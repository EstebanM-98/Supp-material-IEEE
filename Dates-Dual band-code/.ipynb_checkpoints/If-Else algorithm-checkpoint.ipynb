```python
#Library

import numpy as np
from scipy.optimize import curve_fit
import pandas as pd
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
import matplotlib
from sklearn.metrics import r2_score

```

# Functions


```python

%matplotlib inline
```

```python
#Spectral width - 3dB
def Delta_landa(y_data,x_data,err):
    cs=CubicSpline(x_data,y_data)
    new_xdata=np.linspace(min(x_data),max(x_data),len(x_data)*10)
    Y=cs(new_xdata)
    medium_value=np.max(Y)-3
    M_V=[]
    
    for i in range(len(Y)):
        if Y[i]>medium_value:
            M_V.append(i)
    x1=new_xdata[np.min(M_V)]; 
    x2=new_xdata[np.max(M_V)]

    return abs(x2-x1)

def bombeo2(x):
    C_bombeo1= 0.55579579;C_bombeo2= -30.6020973 
    return C_bombeo1*x+C_bombeo2

#Fit (sech^2(x))
def sech(x,zero,power,x0,w):
    f=np.sqrt(power)*((1/(np.cosh((x-x0)/w)))**2)+zero
    return f
#Filter smothing
def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'same') / w



```

# Fitting


```python
# Decision Algorithm

def If_Else(wavelength, power, plot):
    """
    -------------
    wavelength : Array.
    Describes the wavelength interval of the spectrum.
    --------------
    power : Array.
    Describes the spectral power values associated with wavelength values of wavelength.
    --------------
    plot : Conditional.
    Determines whether to plot the fitted data.
    -------------
    Return : String.
    Outputs a class of spectra.
    """
    
    # Smoothing with a moving window
    power = moving_average(power, 5)
    
    # Different scales
    power_linear = power
    power = 10 * np.log10((abs(power) / 1e-3 + 1e-5))
    
    # Fit parameters
    popt_ajust_1533, pcov_ajust_1533 = curve_fit(sech, wavelength[600:920:2], power[600:920:2], bounds=[[-45, 0, 1531, 0], [-44.5, 1700, 1535, 14]], maxfev=1e5)
    popt_ajust_1555, pcov_ajust_1555 = curve_fit(sech, wavelength[1100:1400:2], power[1100:1400:2], bounds=[[-45, 0, 1554, 0], [-44.5, 1700, 1558, 14]], maxfev=1e5)
    #print(popt_ajust_1533, "1533")
    #print(popt_ajust_1555, "1555")

    # Fitted functions
    y_1533 = sech(wavelength, *popt_ajust_1533)
    y_1555 = sech(wavelength, *popt_ajust_1555)

    # Rename fitting variables
    Width_1533 = popt_ajust_1533[3]
    Amplitude_1533 = popt_ajust_1533[1]
    Width_1555 = popt_ajust_1555[3]
    Amplitude_1555 = popt_ajust_1555[1]

    Width_spectral_3dB_1533 = Delta_landa(y_1533, wavelength, 0.01)
    Width_spectral_3dB_1555 = Delta_landa(y_1555, wavelength, 0.01)

    
    #print("Spectral_width and mode locking central wavelenght in 1556-nm band gain = ", [Width_spectral_3dB_1555, popt_ajust_1555[2]])


    # Boundary conditions on which the code is based to group the data
    lower_limit_1533 = 2.85
    upper_limit_1533 = 14
    Minimum_amplitude_1533 = 630
    lower_limit_1555 = 2.7
    upper_limit_1555 = 14
    Minimum_amplitude_1555 = 630

    # Removing null data
    if popt_ajust_1555[1] < 200:
        Width_1555 = 0
        Width_spectral_3dB_1555 = 0
    if popt_ajust_1533[1] < 200:
        Width_1533 = 0
        Width_spectral_3dB_1533 = 0

   

    # IF-ELSE algorithm
    if (Width_1555 < upper_limit_1555 and Width_spectral_3dB_1555 > 2 and Width_1555 > lower_limit_1555 and Amplitude_1555 > Minimum_amplitude_1555) and (Amplitude_1533 < Minimum_amplitude_1533 or Width_1533 < 2.5):
        class_type = "1555"
    elif (Width_1533 > 2.85 and Width_spectral_3dB_1533 > 2 and Amplitude_1533 > Minimum_amplitude_1533) and (Width_1555 > 2.7 and Amplitude_1555 > Minimum_amplitude_1555):
        class_type = "DB"
    elif (Width_1533 < lower_limit_1533 and Width_1555 < lower_limit_1555) or (Amplitude_1533 < Minimum_amplitude_1533 and Amplitude_1555 < Minimum_amplitude_1555):
        class_type = "CW"
    elif (Width_1533 < upper_limit_1533 and Width_1533 > lower_limit_1533 and Amplitude_1533 > Minimum_amplitude_1533) and (Width_1555 < lower_limit_1555 or Amplitude_1555 < Minimum_amplitude_1555):
        class_type = "1533"
    else:
        class_type= "Unknown"
    

    # Plotting if the PLOT conditional is True
    if plot == True:
        
        #Plotting log date
        plt.figure(0,dpi=200)
        plt.xlim(1520, 1570)
        plt.plot(wavelength, power, "o", markersize=1.5)
        plt.plot(wavelength, power, "--", linewidth=2,label="Experimental date")
        plt.plot(wavelength, y_1533, "x", markersize=1.5,label="Fit 1530")
        plt.plot(wavelength, y_1555, "x", markersize=1.5,label="Fit 1550")
        
        # Fitted data
        plt.xlabel("Wavelength [nm]",fontweight="bold")
        plt.ylabel("Power [dB]",fontweight="bold")
        plt.grid(color='black', linestyle='-', linewidth=.5)
        plt.legend()
        plt.show(True)
                    
        # Printing the class and R^2 coefficients

        # Calculation of the coefficient of determination (R^2)
        r_1533 = r2_score(power[600:920], y_1533[600:920])
        r_1555 = r2_score(power[1100:1400], y_1533[1100:1400])
        
        print("Archive name : ", name)
        print("Class spectrum : ",class_type)
        print("Spectral_width and mode locking central wavelenght in 1530-nm band gain = ", [Width_spectral_3dB_1533, popt_ajust_1533[2]])
    
        # Printing the coefficient of determination
        #print(f"Coefficient of determination (R^2) for 1533: {r_1533}")
        #print(f"Coefficient of determination (R^2) for 1555: {r_1555}")

        # Plotting linear date
        #plt.figure(1)
        #plt.xlim(1525, 1565)

        #plt.plot(wavelength, power_linear, "o", markersize=1.5, label=name)
        #plt.plot(wavelength, power_linear, "--", linewidth=2)
        #plt.plot(wavelength, 1e-3 * ((10 ** ((y_1533) / 10)) - 1e-5), "o", markersize=1.5, label=name)
        #plt.plot(wavelength, 1e-3 * ((10 ** ((y_1555) / 10)) - 1e-5), "--", linewidth=2)

    return class_type
```

The "If-Else" algorithm attached uses 2 fitting parameters directly and one indirectly. It directly employs the spectral width (Δλ) and the square root of the signal amplitude, by setting boundary conditions that allow classifying a spectrum into one or another category. On the other hand, it indirectly utilizes the central wavelength of mode-locked ($λ_c$) by centering the search for adjustments in the region of the erbium gain bands (1530 nm and 1556 nm).

```python

# PLot generation
bombe_arr=np.arange(60,460,10)
Bom=bombeo2(bombe_arr)
fig = plt.figure(figsize=(12,10), dpi=100, layout='constrained')
ax = fig.add_subplot(1, 1, 1, projection='3d')
j = 217  # Initial value for z-axis
label = 16  # Label size for axes

#y_ticks=[1.525,1.537,1.550,1.563,1.575]
for p in range(0,50,1):
    
    name="MUESTRA_"+str(2)+"_"+str(12)+"_"+str(p)+".csv"
    df=pd.read_csv(name,skiprows=27,on_bad_lines='skip')
    landa=df["Stop"]
    landa=landa.to_numpy()
    Pot=df["1590.1"]
    Pot=Pot.to_numpy()
    report=If_Else(landa,Pot,False)
    if report=="1533":
        color="r"
    elif report=="1555":
        color="g"
    elif report=="CW":
        color="b"
    else :
        color="y"
   
    
    #Logarithm scale
    
    #Pot=Pot/max(Pot)
    Pot=10*np.log10((abs(Pot)/1e-3+1e-5))
    
    # ... [Loading and processing data]
    landa=landa*1e-3
    ax.plot(landa,Pot,"-",zs=j,color=color, zdir='y',linewidth=1)
    #ax.set_xlim(min(landa),1.575)
    ax.set_ylim(0,max(Bom))
    ax.set_title("$b)$",fontsize=20,fontweight="bold")
    ax.set_xlabel('Wavelength [um]',size=label,fontweight="bold",labelpad=30)
    ax.set_ylabel('$\phi_{2}$ [$^o$]',size=label,fontweight="bold",labelpad=30)
    ax.set_zlabel('Spectrum power [dBm]',size=label,fontweight="bold",labelpad=30)
    
    # ... [Adjusting labels, ticks, and other plot settings]

    font = {'size': 16}
    ax.tick_params('x', labelsize=font['size'])
    ax.tick_params('y', labelsize=font['size'],direction='out',pad=15)
    ax.tick_params("z",labelsize=font['size'],direction='out',pad=15)
    ax.set_zticks([-45,-20])
    plt.tight_layout
    ax.add_collection3d(plt.fill_between(landa,Pot,np.min(Pot),color=color,alpha=.1), zs=j, zdir='y') # lower boundary z=0
    ax.view_init(80,265,"z")
    j=j-.8*abs(Bom[0:1]-Bom[1:2])
    ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    ax.grid(False)
    
%cd ..
%cd ..
plt.savefig(name+".png",bbox_inches='tight')
```
